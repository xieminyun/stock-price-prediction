```python
import yfinance as yf
from pandas_datareader import data as pdr
from datetime import datetime
import pandas as pd

# Read tickers from file
with open('tickers.txt', 'r') as file:
    lines = file.readlines()

tickers = [line.strip() for line in lines]

yf.pdr_override()

end = datetime.now()

# Define start dates for training and testing sets
start_train = datetime(end.year - 5, end.month, end.day)
end_train = datetime(end.year - 1, end.month, end.day)
start_test = datetime(end.year - 1, end.month, end.day)

# Initialize empty dataframes for train and test sets
train_df = pd.DataFrame()
test_df = pd.DataFrame()

for ticker in tickers:
    try:
        # Download stock data
        stock = yf.download(ticker, start_train, end)

        # Add a ticker column to the dataframe
        stock['Ticker'] = ticker

        # Split the data into training and testing datasets
        stock_train = stock[start_train:end_train]
        stock_test = stock[start_test:end]

        # Append the stock data to the respective dataframes using pd.concat
        train_df = pd.concat([train_df, stock_train])
        test_df = pd.concat([test_df, stock_test])
    except Exception as e:
        print(f"Failed to download data for {ticker}. Reason: {e}")

# Save the dataframes to csv files
train_df.to_csv('trainset.csv')
test_df.to_csv('testset.csv')
```
